```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
df = pd.read_excel('data/KCB TOWERS AND SOUTH PODIUM DATA.xlsx')
```

```python
df.head()
```

```python
print(df.describe())
```

### KCB UPPERHILL

```python
# Visualize data for 'KCB UPPERHILL' branch
kcb_upperhill = df[df['BRANCH'] == 'KCB UPPERHILL']

# Set style
sns.set(style="whitegrid")

# Create a color palette
palette = sns.color_palette("mako_r", len(kcb_upperhill.columns) - 2)

# Create the plot
plt.figure(figsize=(12, 8))
barplot = sns.barplot(data=kcb_upperhill.iloc[:, 2:], palette=palette)  # Exclude 'BRANCH' and 'TOTAL' columns

# Add counts on top of each bar
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.0f'), 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha = 'center', va = 'center', 
                     xytext = (0, 10), 
                     textcoords = 'offset points')

plt.title('Health Metrics for KCB UPPERHILL', fontsize=20)
plt.xticks(rotation=90, fontsize=12)  # Rotate x-axis labels for readability
plt.yticks(fontsize=12)
plt.xlabel(None)  # Remove x-axis label
plt.ylabel('Clients Count', fontsize=15)

# Add legend
total = kcb_upperhill['TOTAL'].values[0]
plt.legend(title=f'TOTAL CLIENTS SEEN: {total}', title_fontsize='13',loc ='upper right', bbox_to_anchor=(1.0, 0.9),
           )

# Remove top and right borders
sns.despine()

plt.show()
```

### Using Ploty

```python
import plotly.express as px

# Visualize data for 'KCB UPPERHILL' branch
kcb_upperhill = df[df['BRANCH'] == 'KCB UPPERHILL']

# Melt DataFrame to have health metrics and counts in separate columns
kcb_upperhill_melt = kcb_upperhill.melt(id_vars=['BRANCH', 'TOTAL'], var_name='Health Metric', value_name='Count')

# Create the plot
fig = px.bar(kcb_upperhill_melt, x='Health Metric', y='Count', color='Health Metric',
             title='Health Metrics for KCB UPPERHILL', labels={'Count': 'Count', 'Health Metric': ''})

# Update layout for a cleaner look
fig.update_layout(showlegend=False, plot_bgcolor='white')
fig.update_xaxes(tickangle=45, title_standoff=25)

fig.show()
```
